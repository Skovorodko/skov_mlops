# Make Machine Learning Model with XGBoost

```python
X = train.drop('Survived', axis=1)
y = train['Survived']

print('Shape of X: ', X.shape)
print('Shape of y: ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)

x_test, x_valid, y_test, y_valid = train_test_split(x_test, y_test, test_size=0.5, random_state=1337)
```

```python
x_train
```

```python
from xgboost import XGBClassifier
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
res = []
max_depth = range(1,10)
n_estimators = [50, 100, 150, 200, 250, 300]
learning_rate = [0.01,.05, 0.1, 0.5, 1]
for i in max_depth:
    for n in n_estimators:
        for m in learning_rate:
            gbm = xgb.XGBClassifier(max_depth=i, n_estimators=n, learning_rate=m).fit(x_train,y_train)
            predictions = gbm.predict(x_valid)
            acc_sore = (accuracy_score(y_valid, predictions))   
            res.append((acc_sore, i, n, m))
            print('accuracy:', acc_sore)
            print('learning_rate', m)
            print('n_estimators', n)
            print('max_depth', i)
            print('')
            print()
```

```python
max(res)
```

```python
preds_class = gbm.predict(x_test)
accuracy_score(y_test, preds_class)
```

**XGBRegressor**

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV


params = {'max_depth': range(1,10),
           'n_estimators': [50, 100, 150, 200, 250, 300]}
model = XGBRegressor()

grid = GridSearchCV(estimator=model, 
                   param_grid=params,
                   scoring='r2', 
                   verbose=1)

grid.fit(x_train, y_train)

print("Best parameters:", grid.best_params_)
print("Best Score (R2): ", (grid.best_score_))
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score, r2_score
xgb_tuned = XGBRegressor(**grid.best_params_).fit(x_train, y_train)
y_pred = grid.predict(x_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier

res = []

min_samples_split = range(2,10)
n_estimators = [50, 100, 150, 200, 250, 300]
min_samples_leaf = range(2,10)

for i in min_samples_split:
    for n in n_estimators:
        for m in min_samples_leaf:
            rf = RandomForestClassifier(criterion='gini', 
                             n_estimators=n,
                             min_samples_split=i,
                             min_samples_leaf=m,
                             max_features='auto',
                             random_state=1,
                             n_jobs=-1)
            rf.fit(x_train, y_train)
            predictions = rf.predict(x_valid)
            acc_sore = (accuracy_score(y_valid, predictions))      
            res.append((acc_sore, i, n, m))
            print('accuracy:', acc_sore)
            print('min_samples_split',i)
            print('n_estimators', n)
            print('min_samples_leaf', m)
            print('')
            print()
            
```
